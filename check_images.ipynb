```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import random
from sklearn.cluster import KMeans
```

```python
test_path = 'data/test/'

for f in os.listdir(test_path):
    if f.endswith('jpg'):
        
        im = plt.imread(test_path + f)
        plt.imshow(im)
        plt.show()
```

```python
seg_path = 'data/segmented/'

for f in os.listdir(test_path):
    if f.endswith('jpg'):
        
        im = plt.imread(test_path + f)
        plt.imshow(im)
        plt.show()
```

```python
test_path = 'data/test/'

for f1, f2 in zip(os.listdir(test_path), os.listdir(seg_path)):
    if f1.endswith('jpg') and f2.endswith('jpg'):
        
        im1 = plt.imread(test_path + f1)
        im2 = plt.imread(seg_path + f2)
        plt.imshow(im1)
        plt.show()
        plt.imshow(im2)
        plt.show()
```
